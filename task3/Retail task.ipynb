# RFM analysis

```python
import datetime as dt
```

```python
df["InvoiceDate"].max()
```

```python
today_date = dt.datetime(2011,12,11)
```

```python
rfm = df.groupby("CustomerID").agg({"InvoiceDate": lambda date : (today_date - date.max()).days,
                                    "InvoiceNo" : "count",
                                    "TotalPrice" : "sum"})
rfm
```

```python
rfm.columns = ["Recency","Frequency","Monetary"]
rfm
```

```python
rfm.describe().T
```

```python
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
```

```python
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method="first"), 5, labels=[1,2,3,4,5])
```

```python
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
rfm['RFM_Segment'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm['RFM_Score'] = rfm[['R_score','F_score','M_score']].astype(int).sum(axis=1)
```

```python
def rfm_segment(score):
    if score >= 13:
        return "Champions"
    elif score >= 10:
        return "Loyal Customers"
    elif score >= 7:
        return "Potential Loyalist"
    elif score >= 5:
        return "At Risk"
    else:
        return "Hibernating"

rfm['Segment'] = rfm['RFM_Score'].apply(rfm_segment)
print(rfm[['Recency','Frequency','Monetary','RFM_Score','Segment']].head())
```

```python
rfm.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=rfm, x="Segment", order=rfm['Segment'].value_counts().index, palette="viridis")
plt.title("Customer Segments (RFM)")
plt.xticks(rotation=45)
plt.show()
```

```python
rfm_pivot_m = rfm.pivot_table(index='F_score', columns='R_score', values='Monetary', aggfunc='mean')

plt.figure(figsize=(8,6))
sns.heatmap(rfm_pivot_m, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap of Average Monetary Value by R & F")
plt.xlabel("Recency Score")
plt.ylabel("Frequency Score")
plt.show()
```

## Marketing Ideas per RFM Segment

- **Champions**:  
  - Give early access to new products  
  - Exclusive discounts or VIP programs  
  - Request reviews and referrals  

- **Loyal Customers**:  
  - Provide personalized recommendations  
  - Offer bundled products or subscription plans  
  - Reward with loyalty points  

- **Potential Loyalist**:  
  - Send limited-time discounts to encourage repeat purchase  
  - Offer free shipping for next order  
  - Nurture into loyal segment with engagement emails  

- **At Risk**:  
  - Send re-engagement campaigns ("We miss you" emails)  
  - Special discounts to bring them back  
  - Survey to understand why they stopped purchasing  

- **Hibernating**:  
  - Send win-back offers with strong incentives (e.g., big discount)  
  - Highlight trending products they might like  
  - Reduce marketing costs if response is low
